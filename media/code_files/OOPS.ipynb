```python
class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number  # Encapsulated attribute (protected)
        self._balance = balance                # Encapsulated attribute (protected)
        self.__transaction_history = []        # Encapsulated attribute (private)

    def deposit(self, amount):
        self._balance += amount
        self.__transaction_history.append(f"Deposited {amount}")

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            self.__transaction_history.append(f"Withdrew {amount}")
        else:
            print("Insufficient balance!")

    def get_balance(self):
        return self._balance

    def get_transaction_history(self):
        return self.__transaction_history


# Creating an instance of the BankAccount class
account1 = BankAccount("123456", 1000)

# Accessing and modifying encapsulated attributes using public methods
print("Account Number:", account1._account_number)  # Correct (Not recommended but possible)
print("Initial Balance:", account1._balance)        # Correct (Not recommended but possible)

# Accessing private attribute directly (Not recommended)
'''print("Transaction History:", account1.__transaction_history)'''  # AttributeError: 'BankAccount' object has no attribute '__transaction_history'

# Accessing private attribute using public method
print("Transaction History:", account1.get_transaction_history())  # Correct

# Modifying encapsulated attribute directly (Not recommended)
account1._balance += 500  # Correct (Not recommended but possible)
print("Modified Balance:", account1._balance)

# Modifying private attribute directly (Not recommended)
# account1.__transaction_history.append("Additional transaction")  # AttributeError: 'BankAccount' object has no attribute '__transaction_history'

# Modifying private attribute using public method
account1.deposit(500)  # Correct
print("Updated Balance:", account1.get_balance())  # Correct
print("Updated Transaction History:", account1.get_transaction_history())  # Correct
```

```python
class QuestionPaperAnalyzer:
    """Analyzes a question paper based on user input."""

    def __init__(self):
        """Initializes the analyzer."""
        self.marks = {}
        self.stats = {}

    def get_marks(self):
        """Gets marks from the user and stores them in the marks dictionary."""
        while True:
            try:
                question_num = int(input("Enter question number (or -1 to finish): "))
                if question_num == -1:
                    break

                mark = float(input(f"Enter marks for question {question_num}: "))
                self.marks[question_num] = mark
            except ValueError:
                print("Invalid input. Please enter a number.")

    def calculate_statistics(self):
        """Calculates statistics based on the stored marks."""
        if not self.marks:
            return

        total_marks = sum(self.marks.values())
        num_questions = len(self.marks)
        average_mark = total_marks / num_questions

        # Find the question with the highest and lowest marks
        highest_mark = max(self.marks.values())
        lowest_mark = min(self.marks.values())
        highest_question = [q for q, m in self.marks.items() if m == highest_mark][0]
        lowest_question = [q for q, m in self.marks.items() if m == lowest_mark][0]

        self.stats = {
            "Total Marks": total_marks,
            "Average Mark": average_mark,
            "Highest Mark": (highest_mark, highest_question),
            "Lowest Mark": (lowest_mark, lowest_question),
        }

    def print_results(self):
        """Prints the marks and statistics."""
        print("Question\tMarks")
        print("-------\t-----")
        for q, m in self.marks.items():
            print(f"{q}\t\t{m:.2f}")

        if self.stats:
            print("\nStatistics:")
            for stat_name, stat_value in self.stats.items():
                if isinstance(stat_value, tuple):
                    value, question = stat_value
                    print(f"{stat_name}: {value:.2f} (Question {question})")
                else:
                    print(f"{stat_name}: {stat_value:.2f}")

if __name__ == "__main__":
    analyzer = QuestionPaperAnalyzer()
    analyzer.get_marks()
    analyzer.calculate_statistics()
    analyzer.print_results()
```
